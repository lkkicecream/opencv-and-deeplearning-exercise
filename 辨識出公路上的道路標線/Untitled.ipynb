```python
# 灰階處理（？
image_gray = mping.imread('23.jpg')
```

```python
#高斯模糊
blur = cv2.GaussianBlur(image_gray,(5,5),0)
plt.imshow(blur)
```

```python
'''# 邊緣檢測
sobelx = cv2.Sobel(blur,cv2.CV_64F,1,0,ksize=3)
sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.Sobel(blur,cv2.CV_64F,0,1,ksize=3)
sobely = cv2.convertScaleAbs(sobely)
sobelxy = cv2.addWeighted(sobelx,0.5,sobely,0.5,0)
plt.imshow(sobelxy)'''
low_threshold = 50
high_threshold = 150
masked_edges = cv2.Canny(blur, low_threshold, high_threshold)
plt.imshow(masked_edges)
```

```python
#ROI，直接用座標 BAD
b  = np.array([[[300,250],  [0,450], [620,450],[350,250]]], dtype = np.int32)
im = np.zeros(masked_edges.shape[:2], dtype = "uint8")
cv2.polylines(im, b, 1, 255)
cv2.fillPoly(im, b, 255)

mask = im
plt.imshow(mask)
masked = cv2.bitwise_and(masked_edges, masked_edges, mask = mask)
plt.imshow(masked)
```
